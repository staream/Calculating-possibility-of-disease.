# 가슴의 병을 나타내기 위해서 이렇게 나타냄

#### 내가 해야 할 것 일단 질병에 대한 표를 만들면 좋겠다.
#### 각 표를 만들어서 입력하면 각 표를 가져오는 거지
#### 거기서 질병을 적어보고, 그것을 통해 결과를 가져오는 것

```python
from bs4 import BeautifulSoup
import requests
a=1# 가슴의 페이지를 나타내고 있다.1~8까지 있다.
last_text_list=[]
chest_symptoms_all=dict()
symptoms_all=[]
for i in range(1,9):
    url=f"https://www.amc.seoul.kr/asan/healthinfo/disease/diseaseList.do?pageIndex={i}&partId=B000020&diseaseKindId=&searchKeyword=%EA%B0%80%EC%8A%B4"
    response = requests.get(url)
    html = response.text
    soup=BeautifulSoup(html,'html.parser')
    a_1= soup.select('#listForm > div > div > ul > li > div.contBox > strong > a')#질병
    text_list=[element.get_text() for element in a_1]
    count=len(text_list)
    for i in text_list:
        last_text_list.append(i)
        symptoms_all.append(i)
    for a in range(0,count):
        a_증상= soup.select(f'#listForm > div > div > ul > li:nth-child({a+1}) > div.contBox > dl > dd:nth-child(2) > a')#증상
        # 윗줄은 증상을 list형으로 해서 증상에 넣고 있다.['하나']
        text_list_1=[element.get_text() for element in a_증상]
        # 거기의 하나의 것을 text_list_1에 넣고 있다.
        chest_symptoms_all[text_list[a]]=text_list_1
        #증상을 dict에 넣는다.
    
#print(last_text_list)
#count=len(last_text_list)
#print(count)
#print(chest_symptoms_all)# 질병에 따른 질환을 사전형에 저장했다.
# 각각의 dict 마지막에 1을 넣을 것이다.

#chest_symptoms_all
```

```python
#관련 질병의 중복개수를 구한다.
# chest_symptoms_all
#symptoms_test=dict()
chest_symptoms_test=dict()
for i in chest_symptoms_all.keys():
    for j in chest_symptoms_all[i]:
        if j not in chest_symptoms_test.keys():#symptoms_test:
            chest_symptoms_test[j]=1#symptoms_test[j]=1
        else:
            chest_symptoms_test[j]+=1#symptoms_test[j]+=1
  
```

```python
real_count=len(chest_symptoms_test)
#질병의 확률을 구한다.
# chest_symptoms_test
point_of_disease=dict()
for i in chest_symptoms_test.keys():
    point_of_disease[i]=(real_count-10-chest_symptoms_test[i])/real_count*100
#point_of_disease
#real_count
#chest_symptoms_test
```
